```python
aar=2021
#tester input
while True:
    try:
       alder=int(input('Hvilket år er du født? '))
    except ValueError:
        continue
        
    else:
        break
#tester logikk i input og skriver ut alder om input er i fortid- om er negativ verdi(f.kr) så skifter det fortegn   
if alder>0 and alder<=aar:   
    print("din alder er {0} år".format(aar-alder))    
elif alder>aar:
    print("du er ikke født enda")
elif alder<0:
     print("din alder er {0} år".format(abs(aar-alder)))
```

```python
import numpy as np
#funksjon for innput
def inndata(var):

    while True:
        try:
           ut = int(input(f'skriv inn {var} i figuren over:'))
        except ValueError:
            continue
        
        else:
            return ut
a=inndata("a")
b=inndata("b")
#regner omkrets som er lengden av b+lengden av c+halve omkretsen av en sirkel med diameter a
omkrets=b+np.sqrt(a**2+b**2)+(np.pi*a)/2
#regnar arealet av en trekant med grunnflate b og høyde a og arealet av en halvsirkel med radius a/2
areal=(b*a/2)+(np.pi*((a/2)**2)/2)

print(f"figuren har ytreomkrets {omkrets} og arealet {areal}")
```

```python
import numpy as np

xstart=0
ystart=0

totalavstand=[]
var=0
counter=1            
luftlinjeavstand=0
xarr=np.array([0])
yarr=np.array([0])
#funksjonen reise som tar innput for en destinasjon x og y koordinat og beregner avstand fra forrige destinasjon
def reise(a):
    while True:
        try:
            xend = float(input(f'x-koord for destinasjon nr.{a}: '))
            yend = float(input(f'y-koord for destinasjon nr.{a}: '))
        except ValueError:
            continue
        
        else:
            break
    distance = np.sqrt((xend-xstart)**2 + (yend-ystart)**2)
    return([distance,xend,yend])
#for løkke som kjører for fire destinasjoner og lagrer avstand til listen totalavstand    
for i in range(4):
    var=reise(counter)
    totalavstand.append(var[0])
    xarr = np.append(xarr,var[1])
    yarr = np.append(yarr,var[2])
    xstart=var[1]
    ystart=var[2]
    counter+=1

#skriver ut summen av alle avstander i listen totalavstand    
print(f"total avstanden langs reiseruten er {sum(totalavstand)}")
#beregner avstanden fra origo og til destionasjon 4 og skriver denne ut.
print(f"lengde i luftlinje er {np.sqrt((xstart)**2 + (ystart)**2)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

#teller for destinasjoner
i=0
#fortegn som brukes til og plasserer diestinasjon annotasjon litt tilfeldig for å minimere at 2 annotasjoner skrives overhverandre
fortegn=[1,-1]
# stiller størrelsen på plottet
plt.figure(figsize=(12, 10), dpi=80)
#plotter reise
plt.plot(xarr,yarr, marker="o", ms=10)
#plotter luftlinje
plt.plot([xarr[0],xarr[-1]],[yarr[0],yarr[-1]])
#plotter metadata til plottet
plt.grid()
plt.xlabel("x",fontsize=10)
plt.ylabel("y",fontsize=10)
plt.legend(("Reise","Luftlinje"), loc="best",fontsize=15)

#for løkke som tegner annotasjon til destinasjonene med Tekst korrdinater og pil til punktet
for x,y in zip(xarr,yarr):
    if i>0:
        plt.annotate(f"destinasjon nr.{i} \n ({x},{y})",(x,y),fontsize=15, xytext=(50*random.choice(fortegn), 50*random.choice(fortegn)), textcoords="offset pixels",arrowprops=dict(facecolor='black', shrink=0.05))
    i+=1

```
